## Extra and Join

```python
extra = copper.read_csv('test/extra.csv')
```

```python
extra
```

```python
reviews = copper.read_csv('test/review.csv')
```

```python
reviews
```

```python
del reviews['type']
```

```python
del reviews['text']
```

```python
everything = pd.merge(reviews, users, on='user_id', how='outer')
```

```python
everything = pd.merge(everything, business, on='business_id', how='outer')
```

```python
everything = everything.join(extra, how='outer')
```

```python
everything
```

```python
ds = copper.Dataset(everything)
```

```python
ds.role[['business_id', 'user_id', 'review_id']] = ds.ID
```

```python
ds.fillna()
```

```python
ds.percent_missing()
```

```python
ds.metadata
```

```python
ds.save('test.ds')
```
